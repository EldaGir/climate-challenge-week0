```python
## Missing values analysis
#-No missing values were found in any of the columns.
# - This indicates that the dataset is complete and does not require imputation or removal of missing data.
```

```python
cols = ["T2M", "T2M_MAX", "T2M_MIN", "PRECTOTCORR", "RH2M", "WS2M", "WS2M_MAX"]
```

```python
from scipy.stats import zscore
z_scores = df[cols].apply(zscore)
```

```python
outliers = (np.abs(z_scores)>3)
outlier_count = outliers.sum().sum()
outlier_count
```

```python
rows_with_outliers = outliers.any(axis=1)
rows_with_outliers.sum()
```

```python
## Outlier Detection and Handling
# z-scores were compute for key climate variables to identify extreme values
# a threshold of |Z| > 3 was used to detect outliers
# number of rows containing outliers: 132
# - Decision:
#   Outliers were retained rather than removed.
#
# - Justification:
#   In climate data, extreme values such as high temperatures or heavy rainfall events are meaningful and represent real-world phenomena rather than errors. Removing them could distort the analysis and eliminate important climate patterns.
#
# - Therefore, all outliers were kept for further analysis.
```

```python
# no missing values were found in the dataset. Therefore, no forward-fill was required and no rows needed to be dropped based on missing value thresholds.
```

```python
monthly_temp = df.groupby("Month")["T2M"].mean()
monthly_temp.plot(kind="line", marker="o", title="Monthly Average Temperature (T2M)")
```

```python
max_month = monthly_temp.idxmax()
min_month = monthly_temp.idxmin()

print("Warmest month:", max_month)
print("Coolest month:", min_month)
```

```python
monthly_rain = df.groupby("Month")["PRECTOTCORR"].sum()
monthly_rain.plot(kind="bar", title="Monthly Total Precipitation")
```

```python
peak_rain_month = monthly_rain.idxmax()
print("Peak rainy month:", peak_rain_month)
```

```python
## Time Series Analysis
#
# - The monthly average temperature shows a clear seasonal pattern, with warmer and cooler periods throughout the year.
#
# - The warmest month is April, while the coolest month is November.
#
# - Rainfall patterns indicate that precipitation is concentrated in specific months, suggesting a defined rainy season.
#
# - The peak rainfall occurs in August followed by July and September, which likely corresponds to Ethiopia’s main rainy period.
#
# - These trends reflect seasonal climate variation and are important for understanding agricultural and environmental conditions.
```
