```python
import pandas as pd
```

```python
import numpy as np
df = pd.read_csv("../data/ethiopia.csv")
df.head()
```

```python
df.info()
df.columns
```

```python
df["Country"] = "Ethiopia"
```

```python
df["Date"] = pd.to_datetime(df["YEAR"] * 1000 + df["DOY"], format="%Y%j")
```

```python
df.head()
```

```python
df["Month"] = df["Date"].dt.month
```

```python
df[["YEAR", "DOY", "Date", "Month"]].head()
```
